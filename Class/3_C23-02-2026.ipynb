```python
mylist = [5,4,6,1,1.7,99.8]
myarray = np.array(mylist)
```

```python
myseries1 = pd.Series(data=mylist)
print(myseries1)
myseries2 = pd.Series(data=myarray)
print(myseries2)

df = pd.concat([myseries1,myseries2],axis=1)
print(df)
df = pd.concat([myseries1,myseries2],axis=0) # DataFrame - 2D collection of data with index and columns
print(df)
```

```python
mylables = ['first', 'second', 'third', 'fourth']
# myseries3 = pd.Series(data=mylist, index=mylables)
# print(myseries3)
myseries4 = pd.Series(data=myarray, index=mylist)
print(myseries4)
```
