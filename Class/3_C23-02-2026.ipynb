### Boolean Indexing

```python
df3
```

```python
high_salary = df3[df3["Salary"] > 50000]
print("Employees with high salary:\n", high_salary)
```

```python
salary = df3[((df3["Salary"] == 50000) | (df3["City"] == 'New York'))]
print("Employees with salary 50000 or from New York:\n", salary)
```

```python
salary = df3[((df3["Salary"] == 50000) & (df3["City"] == 'New York'))]
salary
```

```python
salary1 = df3[(df3["Age"] != 25) & (df3["Salary"] != 70000)]
salary1
```
