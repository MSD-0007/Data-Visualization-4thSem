## Mathematical Operations

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 3, 4, 5, 6])
print("a+b:", a + b)
print("a-b:", a - b)
print("a*b:", a * b)
print("a/b:", a / b)
print("a//b:", a // b)
print("a**b:", a ** b)
print("a%b:", a % b)
```
